# Single cell

```python
from scipy.constants import minute, milli
import os
%run ../pdipy/core.py

solution_dimensions = {  # https://www.eppendorf.com/product-media/doc/en/90676/Eppendorf_Consumables_Technical-data_Cell-Culture-Plate-24-Well.pdf
    'surface area (m^2)': 16.2*milli,
    'solution_depth (m)': 17*milli,
    'solution_volume (m^3)': None
}

pdi = PDIBacterialPkg(verbose = True, jupyter = True)
pdi.define_system(surface_system = True, individual_cell = True, solution_dimensions = solution_dimensions)
pdi.define_bacterium(bacterial_specie = 'Saureus')
pdi.define_photosensitizer(photosensitizer = 'A3B_4Zn')
pdi.define_light(light_source = 'LED', lux = 26400,)
pdi.singlet_oxygen_calculations(timestep = 10, total_time = 500)
pdi.kinetic_calculation()
processed_data = pdi.export()

%matplotlib inline
```

# Solution system

```python
from scipy.constants import minute, micro, liter
import os
%run ../../pdipy/core.py

solution_dimensions = {  # https://www.eppendorf.com/product-media/doc/en/90676/Eppendorf_Consumables_Technical-data_Cell-Culture-Plate-24-Well.pdf
    'surface_area (m^2)': 16.2*milli,
    'solution_depth (m)': None,
    'solution_volume (m^3)': 200 *micro*liter
}

pdi = PDIBacterialPkg(verbose = True, jupyter = True)
pdi.define_system(timestep = 10, total_time = 500, surface_system = False, solution_dimensions = solution_dimensions)
pdi.define_bacterium(bacterial_specie = 'Saureus')
pdi.define_photosensitizer(photosensitizer_molar = 1*milli, photosensitizer = 'A3B_4Zn')
pdi.define_light(light_source = 'LED', lux = 26400,)
pdi.singlet_oxygen_calculations()
pdi.kinetic_calculation()
processed_data = pdi.export(simulation_name = 'solution_system')

%matplotlib inline
```

# Surface system

```python
from scipy.constants import minute
import os
%run ../../pdipy/core.py
%matplotlib inline

pdi = PDIBacterialPkg(verbose = True, jupyter = True)
pdi.define_system(timestep = 10, total_time = 500,surface_system = True)
pdi.define_bacterium(bacterial_specie = 'Saureus')
pdi.define_photosensitizer(photosensitizer = 'A3B_4Zn')
pdi.define_light(light_source = 'LED', lux = 26400,)
pdi.singlet_oxygen_calculations()
pdi.kinetic_calculation()
processed_data = pdi.export(simulation_name = 'surface_system')

value, unit = pdi.data_processing(target_time = 0.98)
print(value, unit)
```
